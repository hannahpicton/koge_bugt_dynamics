```python
# Import required modules 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import os 
import matplotlib.lines as mlines
from datetime import datetime
from matplotlib.gridspec import GridSpec
from matplotlib.ticker import FuncFormatter
```

```python
# Define the file paths to the ITS LIVE velocity data for Koge Bugt North (Note this data has already been filtered in the script for Figure 2)
KBN_velocity_file_paths = {'T': 'R:/KOGE_BUGT/github/data/ITS_LIVE_velocity_fahnestock_widget/koge_bugt_north/lat_65.20119249_lon_-40.69217321.csv',
    'T3': 'R:/KOGE_BUGT/github/data/ITS_LIVE_velocity_fahnestock_widget/koge_bugt_north/lat_65.22202097_lon_-40.73106687.csv',
    'T6': 'R:/KOGE_BUGT/github/data/ITS_LIVE_velocity_fahnestock_widget/koge_bugt_north/lat_65.24306868_lon_-40.76932717.csv',
    'T9': 'R:/KOGE_BUGT/github/data/ITS_LIVE_velocity_fahnestock_widget/koge_bugt_north/lat_65.26371721_lon_-40.80884196.csv'}

sampling_location_colors = {'T': '#f77f00', 'T3': '#a30b37', 'T6': '#219ebc', 'T9': '#10678e'}

## PLOT FIGURE S3 ##
fig, ax = plt.subplots(figsize=(9, 5))
for location, path in KBN_velocity_file_paths.items():
    df = pd.read_csv(path)
    df['mid_date'] = pd.to_datetime(df['mid_date']).dt.date  
    ax.scatter(df['mid_date'], df[' v [m/yr]']/1000, marker='o', linestyle='None', s=4, color=sampling_location_colors[location], alpha=0.3)
    ax.plot(df['mid_date'], df['rolling_avg']/1000, linestyle='-', label=f'{location}', color=sampling_location_colors[location], linewidth=2)

ax.set_ylabel('Ice velocity (km/yr)', labelpad=15, fontsize=14)
ax.set_xlabel('Year', labelpad=10, fontsize=14)
ax.grid(axis='x', linestyle='--', alpha=0.9)
ax.set_xlim(pd.to_datetime('2016-01-01'), pd.to_datetime('2024-01-01'))
ax.tick_params(axis='y', labelsize=13)
ax.tick_params(axis='x', labelsize=13)
ax.set_ylim(0, 4.2)
ax.legend(loc='upper left', bbox_to_anchor=(0.035, 0.985), ncol=4, frameon=True, framealpha=1, fontsize=13, markerscale=1, columnspacing=0.8, handletextpad=0.5, fancybox=True, borderaxespad=0.8, edgecolor='gray', handlelength=1.1)
plt.tight_layout()
#plt.savefig('R:/KOGE_BUGT/github/figures/Figure_S3.png', dpi=300, bbox_inches='tight')
plt.show()
```
