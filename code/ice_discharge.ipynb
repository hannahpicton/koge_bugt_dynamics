```python
# IMPORT MODULES
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
from matplotlib.gridspec import GridSpec
import numpy as np 
import datetime
```

## Koge Bugt C

```python
# Plot a timeseries of ice discharge from Koge Bugt C between 2016 and 2024
koge_bugt_c = pd.read_csv('R:/KOGE_BUGT/github/data/mankoff_2020_ice_discharge/koge_bugt_C_ice_discharge.csv')
koge_bugt_c['Date'] = pd.to_datetime(koge_bugt_c['Date'])

plt.rcParams["font.family"] = "Arial"
plt.figure(figsize=(8, 3.5))
plt.plot(koge_bugt_c['Date'], koge_bugt_c['Discharge'], color='#009fb7', linewidth=1.5)
plt.ylabel('Ice Discharge (Gt/yr)', labelpad=10, fontsize=14)
plt.xlabel('Year', labelpad=10, fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.9)
plt.xlim(pd.to_datetime('2016-01-01'), pd.to_datetime('2024-01-01'))
plt.axhline(15.93, color='#009fb7', linestyle='--', linewidth=1.2, alpha=0.6) # Add a horizontal line for the mean ice discharge (2016-2024)
plt.tick_params(axis='y', labelsize=13)
plt.tick_params(axis='x', labelsize=13)
plt.ylim(12.5, 18.5)
plt.tight_layout()
plt.show() 
```

## Koge Bugt S

```python
# Plot a timeseries of ice discharge from Koge Bugt S between 2016 and 2024
koge_bugt_s = pd.read_csv('R:/KOGE_BUGT/github/data/mankoff_2020_ice_discharge/koge_bugt_S_ice_discharge.csv')
koge_bugt_s['Date'] = pd.to_datetime(koge_bugt_s['Date'])
plt.rcParams["font.family"] = "Arial"
plt.figure(figsize=(8, 3.5))
plt.plot(koge_bugt_s['Date'], koge_bugt_s['Discharge'], color='#022b3a', linewidth=1.5)
plt.ylabel('Ice Discharge (Gt/yr)', labelpad=10, fontsize=14)
plt.xlabel('Year', labelpad=10, fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.9)
plt.xlim(pd.to_datetime('2016-01-01'), pd.to_datetime('2024-01-01'))
plt.tick_params(axis='y', labelsize=13)
plt.tick_params(axis='x', labelsize=13)
plt.axhline(8.17, color='#022b3a', linestyle='--', linewidth=1.2, alpha=0.6) # Add a horizontal line for the mean ice discharge (2016-2024)
plt.ylim(7.4, 8.9)
plt.tight_layout()
plt.show() 
```

## Koge Bugt N

```python
# Plot a timeseries of ice discharge from Koge Bugt N between 2016 and 2024
koge_bugt_n = pd.read_csv('R:/KOGE_BUGT/github/data/mankoff_2020_ice_discharge/koge_bugt_N_ice_discharge.csv')
koge_bugt_n['Date'] = pd.to_datetime(koge_bugt_n['Date'])
plt.rcParams["font.family"] = "Arial"
plt.figure(figsize=(8, 3.5))
plt.plot(koge_bugt_n['Date'], koge_bugt_n['Discharge'], color='#f46036', linewidth=1.5)
plt.ylabel('Ice Discharge (Gt/yr)', labelpad=10, fontsize=14)
plt.xlabel('Year', labelpad=10, fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.9)
plt.xlim(pd.to_datetime('2016-01-01'), pd.to_datetime('2024-01-01'))
plt.tick_params(axis='y', labelsize=13)
plt.tick_params(axis='x', labelsize=13)
plt.axhline(4.17, color='#f46036', linestyle='--', linewidth=1.2, alpha=0.6) # Add a horizontal line for the mean ice discharge (2016-2024)
plt.ylim(3.4, 4.9)
plt.tight_layout()
plt.show() 
```

```python
# Combine the Koge Bugt N, C, and S ice discharge timeseries into a single figure with three subplots
fig = plt.figure(figsize=(8.5, 7))
gs = GridSpec(3, 1, height_ratios=[1, 1.4, 1])

# KOGE BUGT N 
ax0 = fig.add_subplot(gs[0])
ax0.plot(koge_bugt_n['Date'], koge_bugt_n['Discharge'], color='#f46036', linewidth=1.5)
ax0.set_ylabel('Ice Discharge (Gt/yr)', labelpad=10, fontsize=13)
ax0.grid(axis='x', linestyle='-', alpha=0.9)
ax0.set_xlim(pd.to_datetime('2016-01-01'), pd.to_datetime('2024-01-01'))
ax0.tick_params(axis='y', labelsize=13)
ax0.tick_params(axis='x', labelsize=13)
ax0.set_ylim(3.3, 5.2)
ax0.tick_params(labelbottom=False)
ax0.text(0.03, 0.90, 'Koge Bugt N', transform=ax0.transAxes, fontsize=15, fontweight='bold', color='#f46036', va='top', ha='left')
ax0.axhline(4.17, color='#f46036', linestyle='--', linewidth=1.2, alpha=0.6) # Add a horizontal line for the mean ice discharge (2016-2024)

# KOGE BUGT C 
ax1 = fig.add_subplot(gs[1])
ax1.plot(koge_bugt_c['Date'], koge_bugt_c['Discharge'], color='#009fb7', linewidth=1.5)
ax1.set_ylabel('Ice Discharge (Gt/yr)', labelpad=14, fontsize=13)
ax1.grid(axis='x', linestyle='-', alpha=0.9)
ax1.set_xlim(pd.to_datetime('2016-01-01'), pd.to_datetime('2024-01-01'))
ax1.tick_params(axis='y', labelsize=13)
ax1.tick_params(axis='x', labelsize=13)
ax1.set_ylim(12.5, 18.5)
ax1.tick_params(labelbottom=False)
ax1.text(0.03, 0.92, 'Koge Bugt C', transform=ax1.transAxes, fontsize=15, fontweight='bold', color='#009fb7', va='top', ha='left')
ax1.axhline(15.93, color='#009fb7', linestyle='--', linewidth=1.3, alpha=0.6) # Add a horizontal line for the mean ice discharge (2016-2024)

# KOGE BUGT S 
ax2 = fig.add_subplot(gs[2])
ax2.plot(koge_bugt_s['Date'], koge_bugt_s['Discharge'], color='#022b3a', linewidth=1.5)
ax2.set_ylabel('Ice Discharge (Gt/yr)', labelpad=10, fontsize=13)
ax2.set_xlabel('Year', labelpad=10, fontsize=14)
ax2.grid(axis='x', linestyle='-', alpha=0.9)
ax2.set_xlim(pd.to_datetime('2016-01-01'), pd.to_datetime('2024-01-01'))
ax2.tick_params(axis='y', labelsize=13)
ax2.tick_params(axis='x', labelsize=13)
ax2.set_ylim(7.3, 9.2)
ax2.text(0.03, 0.90, 'Koge Bugt S', transform=ax2.transAxes, fontsize=15, fontweight='bold', color='#022b3a', va='top', ha='left')
ax2.axhline(8.17, color='#022b3a', linestyle='--', linewidth=1.3, alpha=0.6) # Add a horizontal line for the mean ice discharge (2016-2024)
plt.tight_layout()
plt.show()
```
